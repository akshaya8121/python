```python
#import the libraries
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('/content/Wholesale customers data.csv')
df.head()
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
df.drop(['Channel','Region'],axis=1,inplace=True)
```

```python
array=df.values
```

```python
array
```

```python

stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(20,6))
dendo = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customer data')
plt.ylabel('Eucl distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
group = AgglomerativeClustering(n_clusters=3)
group.fit_predict(X)
```

```python
cluster = group.fit_predict(X)
cluster.shape
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(X,cluster)
```

```python
X
```

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
wcss = []
for i in range(2,11):
  kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#fittting k-means to the dataset
model = KMeans(n_clusters=5,random_state=363)
groups = model.fit_predict(X)
```

```python
groups
```

```python
group_num=pd.DataFrame(groups,columns=['Group'])
group_num
```

```python
cust_kmeans_data = pd.concat([df,group_num],axis=1)
cust_kmeans_data
```

```python
cust_kmeans_data[cust_kmeans_data['Group']==4]
```
