```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import zscore

data=sns.load_dataset('iris')
df=data.copy()
```

```python
z_scores = np.abs(zscore(df.drop('species',axis=1)))
z_scores
```

```python
np.abs(zscore(df.drop('species',axis=1)))
```

```python
non_outliers = (z_scores <= 3).all(axis=1)
df_no_outliers = df[non_outliers]
```

```python
df_no_outliers
```

```python
outliers = (z_scores > 3).any(axis=1)
outlier_rows = df[outliers]
outlier_rows
```

```python
from sklearn.ensemble import IsolationForest
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data=sns.load_dataset('iris')
data
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
data['species']=lb.fit_transform(data['species'])
```

```python
data.head(3)
```

```python
clf = IsolationForest(random_state=0)
clf.fit(data)
```

```python
clf.predict(data)
```

```python
y_pred_outliers = clf.predict(data)
y_pred_outliers
```

```python
idx=np.where(y_pred_outliers==-1)
idx[0]
```
